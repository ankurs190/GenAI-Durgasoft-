- What is Generative AI : It is a branch of artificial intelligence that focuses on creating new content — such as text, images, audio, code, or even 3D models — instead of just analyzing existing data.
- It’s called generative because it generates new outputs, often resembling things a human might create.
- Generative AI regenerates new content by learning patterns from existing data.
- Generative AI is very sensitive toward the input provided in the form of prompt

#### **Session 94 (8th Aug) - Generative AI Intro Cont**

    Large Language Models (LLMs) :
    - LLMs are a type of artificial intelligence model which is trained with large volume data. These models can perform multi task by generating new content.
    - LLMs are build on deep learning and reinforcement learning techniques. These models can understand and recognize the patterns from the input data and can generate human like content based on prompt.
    - Core Components of LLMs :
      - Deep Learning & Reinforcement Learning :LLM's are subset of deep learning and reinforcement learning which are a branch of Machine Learning that uses Artificial Neural Network.
      - Transformers : Transformers are heart of LLMs. These can be considered as core fundamental Architecture of LLMs.
      - Massive Data : LLM's are trained with humongous data . LLM's are designed to perform a specific task for which they are trained with massive datasets.
      - Tokenization : Prior to model processing the test, it has to broken down in to smaller units which are referred tokens.
      - Embeddings : Converting tokens (smaller text units) into numerical representation in the form dense vector is called Embedding. These embedding helps to capture the semantic meaning of token allowing the model to understand relationship between words in a sequence.
      -  Attention Mechanism / Self - Attention Mechanism : This is a very import component of transformer architecture which enables model to concentrate on most relevant parts of the input sequence when generating output
      - Parameter Tuning : LLM's has a very large volume of parameters. These parameters are adjusted during the training process to optimize the model performance.
      - Parameter : These are very important components in the model whose values can changed to enhance performance of the model.

**Mixture of Experts (MoE) LLM**

  - These type of LLMs are designed to perform tasks more efficiently
  - These are combination of small subset language models which are referred as "Experts" (ex: finance+ Image+ Manufacturing + Healthcare)
  - Experts are basically specialized neural networks sub-models which are expertized with specific tasks


| **Aspect**             | **Explanation**                                                                                                   |
| ---------------------- | ----------------------------------------------------------------------------------------------------------------- |
| **Purpose**            | Designed to perform tasks more efficiently by activating only the most relevant parts of the model.               |
| **Structure**          | Composed of multiple smaller sub-models called **"Experts"**.                                                     |
| **Experts**            | Specialized neural network components trained to handle specific types of patterns or tasks.                      |
| **Routing**            | A **Router (Gating Network)** decides which experts to activate for each input or token.                          |
| **Activation Pattern** | Only a small subset of experts (Top-K) are activated per input, making computation **sparse** and cost-effective. |
| **Benefit**            | Can scale parameters into trillions while keeping computational cost manageable.                                  |
| **Example Models**     | Switch Transformer (Google), GLaM (Google), Mixtral (Mistral AI), DeepSpeed-MoE (Microsoft).                      |


**Retrieval Augmented LLM (RAG)**
- RAG = Retrieval + Augmented + Generation
- It’s a hybrid approach where the LLM doesn’t just rely on what it learned during training — it retrieves relevant external knowledge before generating an answer.


      1. Core Idea
      - Traditional LLMs are static — they can only use knowledge from their training data (which gets outdated).
      - RAG models pull in fresh, domain-specific, or private data from an external knowledge base (vector database, search index, documents) in real-time.
      - Then the retrieved context is fed into the LLM’s prompt before generation.

      2. How It Works
      - User Query → Comes in as input.
      - Retrieval Step → Query is converted into an embedding vector and searched in a vector database (e.g., Pinecone, Weaviate, Milvus, FAISS).
      - Context Augmentation → The retrieved documents/snippets are appended to the prompt as extra context.
      - Generation Step → LLM uses the combined input (query + retrieved context) to generate an accurate, up-to-date answer.

      3. Advantages
      - Freshness — Can use the most recent data without retraining.
      - Domain Adaptability — Can work with private/company datasets.
      - Cost Efficiency — No need to retrain the LLM for every data update.
      - Explainability — Can cite sources used for the answer.

      4. Disadvantages
      - Retrieval Quality Dependency — If retrieval fetches irrelevant data, output quality drops.
      - Latency — Retrieval step adds extra processing time.
      - Context Limit — The LLM has a maximum token limit, so you can’t overload it with too much retrieved text.

**Hybrid Architectire LLM**
- Hybrid LLMs combine different architectural types (Encoder-only, Decoder-only, Encoder+Decoder, MoE) or training approaches (RAG, fine-tuning, adapters) in a single system to maximize performance, adaptability, and efficiency.

 **Major Drawbacks of LLM**

    - Data Privacy Risk : If LLM is fine tuned with private data then a strict rules has to be followed in order to avoid data leakage.
    - Resource Intensity : LLM resources intensity refers to how large computational powered demanded for LLM
    - Large Computational Power :
      - Hardware : Training a Large Language Model like Gemini 2.5, GPT 4.0, Llama 3 which consumes thousands of GPU /TPU which may take months to complete training.
      - Energy : LLM training requires large energy (Electricity) Training a GPT - 4 model consumes approx hundreds of megawatt of energy.
      - Memory : LLM requires large memory resources --> RAM 500 - 1000 VRAM in single system.
      - Time : To train LLM with Large volume of data it requires months of time.

  * **Knowledge Limits** – Outdated data, hallucinations, no real understanding.
  * **Bias & Ethics** – Inherits bias, can produce harmful or unbalanced content.
  * **High Cost** – Expensive to train, run, and maintain; high energy use.
  * **Context Limits** – Fixed token window, forgets older context.
  * **Reasoning Gaps** – Weak in complex logic or niche domains.
  * **Security Risks** – Vulnerable to prompt injection, data leakage, misuse.
  * **Lack of Explainability** – Black box behavior, hard to debug.
  * **Overreliance Risk** – Can cause skill loss and spread misinformation.

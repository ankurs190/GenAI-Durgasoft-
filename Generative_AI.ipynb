**Retrieval Augmented LLM (RAG)**
- RAG = Retrieval + Augmented + Generation
- It’s a hybrid approach where the LLM doesn’t just rely on what it learned during training — it retrieves relevant external knowledge before generating an answer.


      1. Core Idea
      - Traditional LLMs are static — they can only use knowledge from their training data (which gets outdated).
      - RAG models pull in fresh, domain-specific, or private data from an external knowledge base (vector database, search index, documents) in real-time.
      - Then the retrieved context is fed into the LLM’s prompt before generation.

      2. How It Works
      - User Query → Comes in as input.
      - Retrieval Step → Query is converted into an embedding vector and searched in a vector database (e.g., Pinecone, Weaviate, Milvus, FAISS).
      - Context Augmentation → The retrieved documents/snippets are appended to the prompt as extra context.
      - Generation Step → LLM uses the combined input (query + retrieved context) to generate an accurate, up-to-date answer.

      3. Advantages
      - Freshness — Can use the most recent data without retraining.
      - Domain Adaptability — Can work with private/company datasets.
      - Cost Efficiency — No need to retrain the LLM for every data update.
      - Explainability — Can cite sources used for the answer.

      4. Disadvantages
      - Retrieval Quality Dependency — If retrieval fetches irrelevant data, output quality drops.
      - Latency — Retrieval step adds extra processing time.
      - Context Limit — The LLM has a maximum token limit, so you can’t overload it with too much retrieved text.

**Evolution of Architectures**

- Traditional ML
  - Couldn’t handle large & multi-dimensional data.
  - Mostly limited to structured data and single-task learning.
- ANN (Artificial Neural Networks)
  - Could process grayscale images.
  - Struggled with very high-dimensional inputs like color images.
- CNN (Convolutional Neural Networks)
  - Specially designed for images (RGB, color).
  - Great for spatial data but weak for sequential/textual data.
- RNN (Recurrent Neural Networks)
  - Designed for sequence/text data.
  - Problem: struggled with long-term dependencies due to vanishing gradients.
- LSTM (Long Short-Term Memory)
  - Improved RNNs by handling long dependencies.
  - Still complex and slow for very large datasets/tasks.
  - Drawback of LSTM- Computational Costly, Complex Architecture, Risk of Over Fitting, Training Time
Interpretability
- Seq2Seq (Encoder-Decoder with LSTMs/GRUs)- https://arxiv.org/pdf/1409.3215
  - Used in machine translation and summarization.
  - Problem: Sequential processing = slow + limited context.
- Attention Mechanism
  - Solved bottleneck of seq2seq by letting model “attend” to relevant parts of the input sequence dynamically.
  - Allowed parallelization instead of step-by-step processing.
  - Major breakthrough toward scalability.
- Self Attention
- Transformers (Attention-Only Models)
  - Fully based on self-attention.
  - Can process massive parallel data, capture long dependencies efficiently
  - Backbone of BERT, GPT, T5, LLaMA, etc.
  - Improved contextual meaning ( better than semantic meaning)
  -  Autoscalability
  - Transfomrer released by Google in 2017 through a paper called  Transformers (2017, "Attention is All You Need")- https://arxiv.org/pdf/1706.03762

  ML → ANN → CNN → RNN → LSTM → Seq2Seq → Attention → Transformers → LLMs.

    - A Transformer is a deep learning architecture designed to process sequential data (like text, speech, or DNA) using the self-attention mechanism, which enables the model to understand contextual meaning and relationships between tokens, regardless of their position in the sequence.
    - Transformers can handle long dependencies and parallel processing
    - Self attention helps us to understand contextual realtionship between different long standing dependencies in a input sequence

    Key Points (for quick recall):
    📌 Parallel processing of sequences (faster than RNN/LSTM).
    📌 Self-Attention → focuses on relevant words in context
    📌 Positional Encoding → keeps track of word order.
    📌 Scalable → foundation of LLMs (BERT, GPT, T5). encoder and decoder are autoscalable. from 6 to many depending on requirement

**Encoder of Transformer**
- it accepts the input sequence, understands it and convert it into set of meaningful vectors which in-turn forwared to decoder for further process
- Core components of encoder
  - Input -> Tokenization + Text Vectorization (Embedding) + Positional Encoding = Input embeddings
  - Mult-head Self Attention
  - Residual Connection (Normalization) and Addition - Add & Norm
  - Feed Forward Neural Network

**Encoder Architecture**
- **Input** – sequential textual data which is going to be processed
  - Example: "Transformers are powerful."
  - Raw text cannot be fed directly → needs preprocessing.
- **Input Embeddings Stage**: This converts raw text into numerical form for the model.
  - **Tokenization**: Breaks the sentence into tokens (words, subwords, or characters). Example (WordPiece/BPE): "Transformers" → "Trans", "former", "s"
  - **Vocabulary Mapping**: extract unique words & providing number represenation to them. Example: "Trans" → 1053, "former" → 8942
  - **Embedding Lookup**: Each ID is mapped to a dense vector based on similarity and semantic meaning (no zeros, sparse vector lot of zeros) (say, 768 dimensions. Example: 1053 → [0.25, -0.67, …, 0.89]

- **Positional Encoding**: It is a technique in transformer model to remember the position of each token in a input sequence.
  - As transformers process input senquence parallely which in-turn results in loosing the actual positions of token which is retained by positional encoder
  -  of two types
    - Absolute- Token is at position n
    - Relative- This token is k steps away from another token

Since Transformers have no recurrence or convolution, they need explicit position info. Adds a sinusoidal or learned vector to embeddings so the model knows word order.

| **Step**                                   | **What Happens**                                                     | **Example with “How are you”**                                                                              |
| ------------------------------------------ | -------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------- |
| **1. Input Text**                          | Raw sequential text enters the model.                                | `"How are you"`                                                                                             |
| **2. Tokenization**                        | Sentence is split into tokens (sub-words/words).                     | `["How", "are", "you"]`                                                                                     |
| **3. Vocabulary Mapping**                  | Tokens mapped to integer IDs from vocabulary.                        | `How → 101`, `are → 202`, `you → 303`                                                                       |
| **4. Word Embeddings**                     | Each token ID converted into dense vector (learned during training). | `How → [0.2, 0.4, 0.6]`<br>`are → [0.5, 0.1, 0.3]`<br>`you → [0.9, 0.7, 0.2]`                               |
| **5. Positional Encoding**                 | Position info added to embeddings (so model knows order).            | `How (pos0) → [0.21, 0.39, 0.64]`<br>`are (pos1) → [0.55, 0.15, 0.32]`<br>`you (pos2) → [0.93, 0.72, 0.25]` |
| **6. Input to Encoder Block**              | Embeddings + positions passed to encoder layer.                      | Matrix of size (3 tokens × embedding\_dim)                                                                  |
| **7. Self-Attention (Scaled Dot Product)** | Each token attends to others → learns context.                       | `"you"` attends strongly to `"are"`, weakly to `"How"`.                                                     |
| **8. Add & Norm**                          | Residual connection + layer normalization.                           | Ensures stability & better gradient flow.                                                                   |
| **9. Feed Forward Network**                | Each token passes through small neural network (non-linear).         | Transforms features for better representation.                                                              |
| **10. Add & Norm (again)**                 | Residual + normalization after FFN.                                  | Final refined embeddings.                                                                                   |
| **11. Encoder Output**                     | Context-aware representation of each token.                          | `"How"` = greeting, `"are"` = verb, `"you"` = subject → all linked semantically.                            |

Self-Attention is a mechanism in Transformers that lets each word in a sentence interact with every other word to capture relationships and context.
It answers:
👉 “When processing this word, which other words in the sentence should I pay attention to?”

self attention derived from attention, attention derived from encoder-decoder

    📖 Mini Example: "How are you"
    Imagine embeddings are very small (just 2D for simplicity).
    Input: ["How", "are", "you"]
    Step:
    "How" creates Q, K, V.
    Q("How") is compared with K("How"), K("are"), K("you").
    Suppose similarity scores = [0.7, 0.2, 0.1].
    After softmax = [0.6, 0.25, 0.15].
    Final representation of "How" =
    0.6×V("How") + 0.25×V("are") + 0.15×V("you").
    So, “How” no longer just means “How” → it now understands its context (“How are you”).
